```python
import matplotlib.pyplot as plt

x = list(range(1, 1000))
y = []

# Calculate the decrement value for linear decrease
decrement = 1 / (len(x) - 1)

# Generate the y values
for i in range(len(x)):
    y.append(1 - i * decrement)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Decrease of y')
plt.show()
```
