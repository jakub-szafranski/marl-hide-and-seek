```python
# Import required libraries
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load data from JSON file
with open("data_test.json", "r") as file:
    data = json.load(file)

from_episode = 9500
# Convert data to DataFrame
df = pd.DataFrame(
    {
        "episode_lengths": data["episode_lengths"][from_episode:],
        "hider_returns": data["hider_returns"][from_episode:],
        "seeker_returns": data["seeker_returns"][from_episode:],
    }
)

# Calculate rolling mean for the last 100 episodes
df["hider_returns_rolling"] = df["hider_returns"].rolling(window=50).mean()
df["seeker_returns_rolling"] = df["seeker_returns"].rolling(window=50).mean()
df["episode_lengths_rolling"] = df["episode_lengths"].rolling(window=50).mean()

# Plot Nagrody Skumulowane (Cumulative Rewards)
plt.figure(figsize=(10, 5))
# plt.plot(df['hider_returns_rolling'], label='Nagrody Skumulowane (Hider)')
plt.plot(df["seeker_returns_rolling"], label="Nagrody Skumulowane Agenta Szukajcego")
plt.xlabel("Rozgrywka")
plt.ylabel("Nagrody Skumulowane")
# plt.title('Nagrody Skumulowane w Czasie')
plt.legend()
plt.grid(True)
plt.show()

# Plot Dugo Rozgrywki (Episode Lengths)
plt.figure(figsize=(10, 5))
plt.plot(df["episode_lengths_rolling"], label="Dugo Rozgrywki", color="orange")
plt.xlabel("Rozgrywka")
plt.ylabel("Dugo Rozgrywki")
# plt.title('Dugo Rozgrywki w Czasie')
plt.legend()
plt.grid(True)
plt.show()

# if episode length == 12 winner == hider alse seeker
print(f"Success rate for hider: {len(df[df['hider_returns'] > 30]) / df.shape[0]}")
print(f"Success rate for seeker: {1 - len(df[df['hider_returns'] > 30]) / df.shape[0]}")
```

```python
# Import required libraries
import json
import pandas as pd
import matplotlib.pyplot as plt

# List of JSON files
json_files = [
    "data_testq2.json",
    "data_testsq3.json",
    "data_testq.json",
    "data_tests.json",
    "data_tests2.json",
    "data_testsm1.json",
    "data_testsm2.json",
    "data_testsm3.json",
    "data_testsm4.json",
]

# Initialize DataFrame to store combined data
combined_df = pd.DataFrame()

for json_file in json_files:
    # Load data from JSON file
    with open(json_file, "r") as file:
        data = json.load(file)

    from_episode = 0
    to_episode = 30000
    # Convert data to DataFrame
    df = pd.DataFrame(
        {
            "episode_lengths": data["episode_lengths"][from_episode:to_episode],
            "hider_returns": data["hider_returns"][from_episode:to_episode],
            "seeker_returns": data["seeker_returns"][from_episode:to_episode],
        }
    )

    # Calculate rolling mean for the last 1000 episodes
    df["hider_returns_rolling"] = df["hider_returns"].rolling(window=1000).mean()
    df["seeker_returns_rolling"] = df["seeker_returns"].rolling(window=1000).mean()
    df["episode_lengths_rolling"] = df["episode_lengths"].rolling(window=1000).mean()

    # Add a column to identify the source file
    df["source"] = json_file

    # Append to combined DataFrame
    combined_df = pd.concat([combined_df, df])

# Define line styles and colors
line_styles = ["-", "--", "-.", ":", "-", "--", "-.", ":", "-"]
colors = ["b", "g", "r", "c", "m", "y", "k", "orange", "purple"]
markers = ["o", "v", "^", "<", ">", "s", "p", "*", "x"]

# Plot Nagrody Skumulowane (Cumulative Rewards)
plt.figure(figsize=(15, 10))
for i, json_file in enumerate(json_files):
    subset = combined_df[combined_df["source"] == json_file]
    plt.plot(
        subset["seeker_returns_rolling"],
        label=f"Nagrody Skumulowane Agenta Szukajcego ({json_file})",
        linestyle=line_styles[i % len(line_styles)],
        color=colors[i % len(colors)],
        marker=markers[i % len(markers)],
        alpha=0.7,
    )
plt.xlabel("Rozgrywka")
plt.ylabel("Nagrody Skumulowane")
plt.legend()
plt.grid(True)
plt.show()

# Plot Dugo Rozgrywki (Episode Lengths)
plt.figure(figsize=(15, 10))
for i, json_file in enumerate(json_files):
    subset = combined_df[combined_df["source"] == json_file]
    plt.plot(
        subset["episode_lengths_rolling"],
        label=f"Dugo Rozgrywki ({json_file})",
        linestyle=line_styles[i % len(line_styles)],
        color=colors[i % len(colors)],
        marker=markers[i % len(markers)],
        alpha=0.7,
    )
plt.xlabel("Rozgrywka")
plt.ylabel("Dugo Rozgrywki")
plt.legend()
plt.grid(True)
plt.show()
```
